```python
def myLineEditor():
    list = ArrayList()
    while True:
        command = input("[i-삽입, d-삭제, r-변경, p-출력, l-파일읽기, s-저장, q-종료] ")
        if command == 'i':
            pos = int(input("삽입할 위치: "))
            str = input("삽입할 내용: ")
            list.insert(pos, str)
        elif command == 'd':
            pos = int(input("삭제할 위치: "))
            list.delete(pos)
        elif command == 'r':
            pos = int(input("변경할 위치: "))
            str = input("변경할 내용: ")
            list.replace(pos, str)
        elif command == 'q':
            return
        elif command == 'p':
            print("Line Editor")
            for line in range(list.size()):
                print("%2d" % line, end='')
                print(list.getEntry(line))
            print()
        elif command == 'l':
            filename = "test.txt"
            with open(filename, 'r') as f:
                lines = f.readlines()
                list.clear()
                for line in lines:
                    list.insert(list.size(), line.rstrip())
        elif command == 's':
            filename = "test.txt"
            with open(filename, 'w') as f:
                for i in range(list.size()):
                    f.write(list.getEntry(i))
```
