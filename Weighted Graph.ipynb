```python
#Kruskal Algorithm
def MST_Kruskal(vertex, adj):
    vsize = len(vertex)
    init_set(vsize)
    eList = []

    for i in range(vsize-1):
        for j in range(i+1, vsize):
            if adj[i][j] != None:
                eList.append((i, j, adj[i][j]))

    eList.sort(key=lambda e: e, reverse=True)
    edgeAccepted = 0

    while edgeAccepted < vsize - 1:
        e = eList.pop(-1)
        uset = find(e)
        vset = find(e)

        if uset != vset:
            print("%s, %s, %d" % (vertex[e], vertex[e], e))
            union(uset, vset)
            edgeAccepted += 1
```

```python
#Prim Algorithm
def MST_Prim(vertex, adj):
    vsize = len(vertex)
    dist = [INF] * vsize
    selected = [False] * vsize
    dist = 0

    for i in range(vsize):
        u = getMinVertex(dist, selected)
        selected[u] = True
        print(vertex[u], end=' ')

        for v in range(vsize):
            if adj[u][v] != None:
                if selected[v] == False and adj[u][v] < dist[v]:
                    dist[v] = adj[u][v]
```

```python
#Dijkstra Algorithm
def shortest_path_dijkstra(vtx, adj, start):
    vsize = len(vtx)
    dist = list(adj[start])
    path = [start] * vsize
    found = [False] * vsize
    found[start] = True
    dist[start] = 0

    for i in range(vsize):
        print("Step%2d" % (i+1), dist)
        u = chooseVertex(dist, found)
        found[u] = True

        for w in range(vsize):
            if not found[w]:
                if dist[u] + adj[u][w] < dist[w]:
                    dist[w] = dist[u] + adj[u][w]
                    path[w] = u
    return path
```
